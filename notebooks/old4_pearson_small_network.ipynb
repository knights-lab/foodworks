```r
library(bnlearn)
library(parallel)

data_numeric = read.table("../data/prediction.small.numeric.txt", header = TRUE, sep="\t", row.names=1, check.names=F, stringsAsFactors=F)
blacklist = read.table("../data/blacklist.small.txt", header=T, sep="\t", row.names=1)
```

```r
data_binary = read.table("../data/prediction.small.binary.txt", header = TRUE, sep="\t", row.names=1, check.names=F, stringsAsFactors=T)
```

```r
data_numeric <- as.data.frame(lapply(data_numeric, as.numeric))
```

```r
# data = cbind(dedup(data_numeric), data_binary)
data = dedup(data_numeric, threshold=.95)
```

```r
# cnames = sample(colnames(data), 100)
cnames = colnames(data)
```

```r
data_small <- data[,cnames]
```

```r
blacklist_small <- blacklist[blacklist$from %in% colnames(data_small),]
blacklist_small <- blacklist_small[blacklist_small$to %in% colnames(data_small),]
```

```r
data_small
```

```r
cl = makeCluster(40, type = "SOCK")
# dag = si.hiton.pc(data_small, cluster = cl, blacklist = blacklist_small, alpha = 0.05, test="mi-cg")
dag = si.hiton.pc(data_small, cluster = cl, blacklist = blacklist_small, alpha=0.05, undirected=F, test="cor")
# dag = si.hiton.pc(data_small, blacklist = blacklist_small, alpha = 0.05)

amat.data = amat(dag)
plot(dag)
write.table(amat.data, "../results/dag.small.all.pearson.txt", sep="\t")
```
